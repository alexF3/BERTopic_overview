## 1. Is it cool?

### Let's say you're on a project for the FDA and they've asked you to help them find trends in research around a specific drug.  
While the agency relies on panels of experts for major regulatory decisions, it wants an algorithmic way of monitoring the literature around a drug or product to augment the situational awareness of senior leaders in the agency.  This can be important in situations in which it becomes challenging to find experts in a medical specialty who are free of relationships with industry that might influence their opinions and approaches.
Purdue Pharma's extended release opioid product Oxycontin provides an excellent case study.  Released in 1996, marketing for Oxycontin focused on the unproven belief that the extended release formula made the drug less likely to cause addiction that other opioids.  Early on, Purdue funded research that significantly influenced the literature on Oxycontin.  Eventually, however, the medical community started to recognize elevated risks of addiction and overdose associated with the use and abuse of the drug.  
<b>The FDA would like you to use the National Library of Medicine's PubMed Platform to download and categorize 3,747 articles on Oxycontin published between 1996 and September 2023 based on latent themes in their abstracts.</b>

There are numerous approaches to this kind of classification.  
* If you had the time, you could manually categorize each document based on its title, but at 20 seconds per classificaiton, but this would take 20 hours of labor and would not scale to other corpera.
* You could spend a fraction of that 20 hours developing a sense for the themes and establishing a set of keywords that might aid in classificaiton, but this would be quite imprecise (what happens if a document has keywords from multiple categories?) and would also not scale very well.
* You could use Latent Derichlet Allocation (LDA), a popular form of NLP that classifies documents based on a "bag of words" approach.  Until recently, this was the state of the art in unsupervised classificaiton with NLP.  One drawback of LDA is that it loses the context of word position and order, which can degrade the informational content of the classifier.
## Or
You could use the BERTopic Python library to leverage pre-trained language models that convert your documents into word embeddings, which are vector representations of the text that retain the contextual information of the words that precede and follow each word in the document.  There are several advantages to this approach:
* <b>BERTopic allows you to use pre-trained language models from [Hugging Face](https://huggingface.co/models) repo to create embeddings of your documents.</b>  This is important because language models are trained on different types of data sources (e.g. Wikipedia, Arxiv papers, or Twitter corpora) and each will perform differently with your application.  Easy integratoion with the Hugging Face API means that work done with BERTopic can evolve over time as new and improved language models become available.
* <b>BERTopic makes it easy to tune hyperparameters in the embedding, dimmension reduction, clustering, vectorizing, and clustering phases of model development. </b> The library aslo provides numerous methods of visualizing the results and fine tuning the final model.


### <b>BERTopic is cool because it lets you use state-of-the-art open source language models to rapidly cluster unstructured text into intuitive and meaningful topic classifications.</b>  
These models often take tens of thousands of hours of compute time on high end GPUs to produce...but you can work with them for free!
